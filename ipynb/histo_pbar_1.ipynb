```python
# linear and logarithmic scale histograms
# 
plt.xscale('log')
plt.yscale('log')

plt.xlabel('Kinetic energy, [GeV]')
plt.ylabel('J [$(\\bar{p}+\\bar{n})$] / J [$(p)$]')
plt.title('Sec. $(\\bar{p}+\\bar{n})$ spectra to prim. $p$ ratio for different models of primary protons spectra')

# WH09: 009*, 010*, 011
WORKDIR = DATADIR+'/010/_analyzed/'
data_x = np.fromfile(WORKDIR+'sec_pbar_sec_nbar_spline_x_log.bin', count=-1, )
data_y = np.fromfile(WORKDIR+'sec_pbar_sec_nbar_spline_y_log.bin', count=-1, )
data_model = np.vectorize(ms.J_WH09)(data_x)
plt.plot(data_x, data_y/data_model, 'g-', label='WH09')

# LA03: 012, 013*
WORKDIR = DATADIR+'/013/_analyzed/'
data_x = np.fromfile(WORKDIR+'sec_pbar_sec_nbar_spline_x_log.bin', count=-1, )
data_y = np.fromfile(WORKDIR+'sec_pbar_sec_nbar_spline_y_log.bin', count=-1, )
data_model = np.vectorize(ms.J_LA03)(data_x)
plt.plot(data_x, data_y/data_model, 'r-', label='LA03')

# BH00: 014*, 015
WORKDIR = DATADIR+'/014/_analyzed/'
data_x = np.fromfile(WORKDIR+'sec_pbar_sec_nbar_spline_x_log.bin', count=-1, )
data_y = np.fromfile(WORKDIR+'sec_pbar_sec_nbar_spline_y_log.bin', count=-1, )
data_model = np.vectorize(ms.J_BH00)(data_x)
plt.plot(data_x, data_y/data_model, 'y-', label='BH00')

# GM75: 016, 017, 018*
WORKDIR = DATADIR+'/018/_analyzed/'
data_x = np.fromfile(WORKDIR+'sec_pbar_sec_nbar_spline_x_log.bin', count=-1, )
data_y = np.fromfile(WORKDIR+'sec_pbar_sec_nbar_spline_y_log.bin', count=-1, )
data_model = np.vectorize(ms.J_GM75)(data_x)
plt.plot(data_x, data_y/data_model, 'c-', label='GM75')

# WH03: 019*, 020**, 021
WORKDIR = DATADIR+'/020/_analyzed/'
data_x = np.fromfile(WORKDIR+'sec_pbar_sec_nbar_spline_x_log.bin', count=-1, )
data_y = np.fromfile(WORKDIR+'sec_pbar_sec_nbar_spline_y_log.bin', count=-1, )
data_model = np.vectorize(ms.J_WH03)(data_x)
plt.plot(data_x, data_y/data_model, 'm-', label='WH03')

plt.legend(loc='best', )
# plt.savefig(DATADIR+'secondary-antiprotons+antineutrons-spectra-to-prim-protons-ratio.svg')
# plt.savefig(DATADIR+'secondary-antiprotons+antineutrons-spectra-to-prim-protons-ratio.png')
plt.show()
```
